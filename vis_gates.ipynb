```python
# Cell 13: random SU2 gate
qc = QuantumCircuit(1)
U = random_su2()                 # 先拿到 2×2 的 SU(2) 矩阵
gate = UnitaryGate(U, label="SU2")
gate._name = "su2"               # 保持和你的传播规则一致
qc.append(gate, [0])
qc.draw('mpl');
```
